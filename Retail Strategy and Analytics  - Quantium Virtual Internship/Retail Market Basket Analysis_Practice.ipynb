```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# Data Preparation

```python
data['Sub-Category'] = data['Sub-Category'].str.strip()
data.dropna(axis= 0, subset=['Order ID'], inplace= True)
data['Order ID'] = data['Order ID'].astype('str')
#data['Order ID'] = data[~data['Order ID'].str.contains('C')]
data.head()
```

```python
my_basket = data.groupby(['Order ID','Sub-Category']).agg({'Quantity':'sum'}).unstack().reset_index().fillna(0).set_index('Order ID')
my_basket.head()
```

```python
def my_encoder(x):
    if x >= 1:
        return 1
    else:
        return 0
    
my_basket_sets = my_basket.applymap(my_encoder)
```

# Lets Train our Model

```python
# Generating frequent itemsets
my_frequent_items = apriori(my_basket_sets, min_support=0.01, use_colnames= True)
my_frequent_items
```

```python
# Generating rules
my_rules = association_rules(my_frequent_items, metric= 'lift', min_threshold= 0.8)
my_rules.sort_values('lift', ascending= False).head(20)
```
